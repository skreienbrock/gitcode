### Suffix Counter ###

```python
import os
import os.path as op
# regex
# import re
# args
import sys

### STATIC VARS
myPFolder='/home/svekre'
```

```python
try: 
    args=sys.argv
    if op.isdir(args[1]):
        maindir=args[1]
    else:
        # fallback
        maindir=myPFolder
except IndexError:
        maindir=myPFolder
```

```python
# DEBUG print
# maindir
```

```python
endings=[]
fCnt=0

for directory,sfolder,files in os.walk(maindir):
    # DEBUG print directories
    #print('\033[1;31m' + directory + '\033[0m')
    
    # DEBUG print subdirectories
    #for s in sfolder:
    #     print('\033[1;32m' + s + '\033[0m')
    
    # but interesting are the files:
    for f in files:
        # dot in files ?
        fCnt = fCnt + 1
        if '.' in f:
            fIndex=(len(f.split('.')))
            endings.append(f.split('.')[fIndex-1].lower())
```

```python
print("In %d files, %d had suffixes." % (fCnt,len(endings)))
```

```python
# all files with endings:
len(endings)
```

```python
# get all endings in folder
endset=set(endings)
# all endings/suffixes
len(endset)
endset
```

```python
# DEBUG Print
# endings.count('json')
```

```python
suffixOccurence={}
for suffix in endset:
    if endings.count(suffix) >= 100:
        suffixOccurence[suffix]=endings.count(suffix)
```

```python
suffixOccurence
```

```python
import operator
sorted_sO = sorted(suffixOccurence.items(), key=operator.itemgetter(1),reverse=True)

# sorted_sO
```
